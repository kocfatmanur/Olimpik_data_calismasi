```python
sport_event_count = dataoa.groupby("Sport")["Event"].nunique().sort_values(ascending = False)
sport_event_count.plot(kind = "bar")
plt.title("Number of Unique Events per sport")
plt.xlabel("Sport")
plt.ylabel("Number of Unique Events")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
year_participant_count =dataoa.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="line",marker="*")
plt.title("Number of Participants Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Participants")
plt.show()
```

```python
year_participant_count =dataoa.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="bar")
plt.title("Number of Participants Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Participants")
plt.show()
```

```python
country_avg_age = dataoa.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(10).plot(kind="bar")
plt.title("Top 10 Countries with Highest Average Age of Participants")
plt.xlabel("Country")
plt.ylabel("Avarege Age")
plt.xticks(rotation=45)
plt.show()
```

```python
country_avg_age = dataoa.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.tail(10).plot(kind="bar")
plt.title("Top 10 Countries with Lowest Average Age of Participants")
plt.xlabel("Country")
plt.ylabel("Avarege Age")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=dataoa, x="Season",y="Age")
plt.title("Distribution of Ages by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data=dataoa,x="Medal", y="Height", palette="Set2")
plt.title("Distribution of Heights by Medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
most_medals_country = dataoa["NOC"].value_counts().idxmax()
print("Most Modal-Wrinning Country:", most_medals_country)
```

```python
talest_athlete = dataoa[dataoa["Height"]==dataoa["Height"].max()]
print("Tallest Athlete:")
print(talest_athlete[["ID","Name","Height","Sport"]])
```

```python
heaviest_athlete = dataoa[dataoa["Weight"]==dataoa["Weight"].min()]
print("Heaviest Athlete:")
print(heaviest_athlete[["ID","Name","Weight","Sport"]])
```

```python
sns.scatterplot(data=dataoa, x= "Height", y="Weight", hue="Medal", palette="plasma")
plt.title("Athlete Height vs Weight by MEdal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(title="Medal")
plt.show()
```
