```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
athletes = pd.read_csv('athlete_events.csv')
regions = pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
athletes_df = athletes.merge(regions, how ='left', on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region': 'Region', 'notes': 'Notes'}, inplace=True)
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Ülkelere Göre Genel Katılım')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set1')
```

```python
plt.figure(figsize=(12,6))
plt.title('Atletlerin Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Kişi Sayısı')
plt.hist(athletes_df.Age, bins=np.arange(10,80,2), color='yellow', edgecolor='black')
```

```python
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Cinsiyet Dağılımı')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=100, shadow=True )
```

```python
athletes_df.Medal.value_counts()
```

```python
Female_participants = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
Female_participants = Female_participants.groupby('Year').count().reset_index()
Female_participants.head()
```

```python
women_olympics = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=women_olympics, palette="Spectral")
plt.title('Katılan Kadın Sayısı')
```

```python
part=women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title("Yıllar İçinde Kadın Sporcu Sayısı")
```

```python
gold_medals = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medals.head()
```

```python
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
sporting_event = gold_medals['Sport'][gold_medals['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(x=sporting_event)
plt.title('60 Yaş Sonrası Altın Madalya Kazanılan Dallar')
```

```python
gold_medals.Region.value_counts().reset_index(name='Medal').head()
```

```python
total_gold_medals = gold_medals.Region.value_counts().reset_index(name='Medal').head()
g = sns.catplot(x="Region", y="Medal", data=total_gold_medals, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("İlk 5 Ülke")
g.set_ylabels("Madalya Sayısı")
plt.title("Ülkere Göre Altın Madalyalar")
```

```python
max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None)
plt.xlabel("2016da En Çok Altın Madalya Kazanan Ülkeler")
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize=(12,10))
axis = sns.scatterplot(x='Height', y='Weight', data=not_null_medals, hue="Sex")
plt.title("Ağırlık vs Kiloya Göre Madalya Kazananlar")
```
